# Data Preprocessing Tools

## Importing the libraries

```
#import libraries by matplot, pandas, and numpy
#sklearn has all Machine learning libraries and packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
#create a new variable for my data set
#read.csv function is inside of pandas library so we have to call pandas first 
#and then call read.csv
dataset= pd.read_csv("Data.csv")
#dependent variable are usually stored in the last column
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(y)
```

## Taking care of missing data

```
#There are several ways to tackle this issue
#1. If we have a large data set and %1 of them is missing we remove those missed data
#2. Replace the missing value by the average of that feature
#we use sklearn to handle missing salary
```

```
#imputer is an object from SimpleImputer class and sklearn is a library
#fit, transform are also a method 
#we need to find all numerical columns to replace any missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy= 'mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
print(X)
```

## Encoding categorical data

```
#Suppose in a large data set we have only 3 categorical data
#so we group them in 3 columns and assign a binary id to them like: 001, 010, ...
```

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
#adding np.array to make sure we convert it to array representation
```

```
print(X)
```

### Encoding the Independent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder ()
y= le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
#which one is correct? feature scaling after splitting dataset or 
#after splitting? Correct answer: AFTER splitting the dataset
#it prevents information leakage for test set
```

```
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test are four variable that will be return by 
#train_test_split class
#test_size =0.2 means %20 of dataset is going for test set and %80 is for training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
#Two approaches for feature scaling:
#1. standardisation: (x- mean(x))/ standard deviation (x)--> good for all kind of data(recommended)
#2. Normalization: (x-min(x))/ (max(x)- min(x))--> good for normal distribution
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#do we have to apply standardisation for dummy nodes as well? NO
#apply standardisation for training set
#we consider new X and y so the indicies are different for age and salary from the original dataset
X_train [:, 3:] = sc.fit_transform(X_train [:, 3:])
#X_train values are between [-2,+2]
X_test [:, 3:] = sc.transform(X_test [:, 3:])
```
